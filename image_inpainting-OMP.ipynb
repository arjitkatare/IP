```python
resi = (200,300)
```

```python
img = cv2.resize(img0,resi)
img1 = cv2.resize(img01,resi)
i1 =  cv2.resize(i01,resi)
i2 = cv2.resize(i02,resi)
i3 =  cv2.resize(i03,resi)
i4 = cv2.resize(i04,resi)
i_main = [img,i1,i2,i3,i4]
```

```python
#Making Basis for Dictionary
size2 = 64
Basis_T = []
w = 4
for i in range(len(i_main)):
    i_temp = i_main[i]
    x,y = i_temp.shape
    B = [0]*64
    for j in range(40000):
        count = 0
        x_1 = random.randint(50,x-50)
        y_1  =random.randint(50,y-50)
        for k in range(-w,w):
            for l in range(-w,w):
                B[count] = i_temp[x_1 + l][y_1 + k]
                count += 1
        Basis_T.append(B/np.linalg.norm(B))
size5 = len(Basis_T)
```

```python
B = []
for i in range(256):
    x = random.randint(0,size5-1)
    B.append(Basis_T[x])
Basis = np.array(np.transpose(B),ndmin = 2)
```

```python
#Making Mask
img_masked = cv2.imread("meno.png",0)
img_mask = cv2.resize(img_masked,resi)
```

```python
mask1 = img_mask
for i in range(len(mask1)):
    for j in range(len(mask1[0])):
        if mask1[i,j] <=253:
            mask1[i,j]=0
```

```python
mask_open = cv2.morphologyEx(mask1,cv2.MORPH_CLOSE,np.ones((3,3),np.uint8))
```

```python
height, width= mask1.shape

for i in range(0, height):
    for j in range(0, width):
        if mask1[i,j] != 0:
            mask1[i,j] = 255
thresh = mask1
count = 0
for i in range(0, height):
    for j in range(0, width):
        if mask1[i,j]!=0:
            count+=1
```

```python
def destroy(inner2,Lambda):
    for i in range(len(inner2)):
        if Lambda[i]!= 0 :
            inner2[i] = 0
    return inner2
```

```python
def boundary5(mask1):
    thresh = mask1
    contours_count = 0
    contours = None
    w = 4
    check = 0
    for i in range(100,len(thresh)-100):
        for j in range(25,len(thresh[0])-25):
            if thresh[i,j]!=0:
                x,y = i,j
                check = 1
                break
        if check == 1:
            break
    if check ==0 :
        return None
    contours = [x-3,y-3]
    return contours
```

```python
t = 0
x,y = [0,0]
while t!=1:
    t+=1
    contours= boundary5(mask1)
    if contours == None:
        break      
    
    w = 4
    patches_img = []
    patches_mask = []
    

    patch_img = []
    patch_mask = []
    x,y = contours
    for i in range(-w,w):
        for j in range(-w,w):
            patch_img.append(img1[x+j,y+i])
            patch_mask.append(mask1[x+j,y+i])
            
    Basis_sub = []
    patch_img_sub = []
    basis_sub = []
    
    for k in range(64):
        if patch_mask[k] <=127:
            patch_img_sub.append(patch_img[k])
            basis_sub.append(Basis[k])
    Basis_sub = np.array(basis_sub)
    #patches_img_sub = patch_img_sub
    
    #size3 = len(patches_img)
    x = np.array(patch_img_sub,ndmin = 2)
    r = np.array(x)

    r_T = r.transpose()
    Phi_T = []
    #m_ind = 0 
    #m_max = 0
    q= 0
    B_T = np.transpose(Basis_sub)
    a_phi = []

    Lambda = np.array([0]*256)
    while q < 20:
        inner2 = np.matmul(B_T,r_T)
        inner2 = destroy(inner2,Lambda)
        inner = inner2.transpose()
        in_max = np.argmax(inner[0])
        Lambda[in_max] = inner[0][in_max]
        Phi_T.append(B_T[in_max])
        nPhi_T = np.array(Phi_T,ndmin=2)
        Phi = np.transpose(nPhi_T)
        x_T = x.transpose()
        a_phi = np.matmul(np.linalg.pinv(Phi),x_T)
        r_T = x_T - np.matmul(Phi,a_phi)
        q+=1
    La = np.array(Lambda,ndmin = 2)
    La_T = La.transpose()
    reco = np.matmul(Basis,La_T)
    reco_T = reco.transpose()
    #print(np.linalg.norm(x),np.linalg.norm(r_T))
    #print(patch_img)
    
    x,y = contours
    count = 0
    for i in range(-w,w):
        for j in range(-w,w):
            if mask1[x+j][y+i]!=0:
                img1[x+j][y+i] = reco_T[0][count]
                mask1[x+j][y+i] = 0
            count+=1
```
