```python
def boundary5(mask1):
    thresh = mask1
    contours_count = 0
    contours = None
    w = 4
    check = 0
    for i in range(100,len(thresh)-100):
        for j in range(25,len(thresh[0])-25):
            if thresh[i,j]!=0:
                x,y = i,j
                check = 1
                break
        if check == 1:
            break
    if check ==0 :
        return None
    contours = [x-3,y-3]
    return contours
```
